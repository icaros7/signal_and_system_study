```python
import numpy as np
import matplotlib.pyplot as plt

# 문제: P1.21
# 학번: 201735030
# 이름: 이호민

t = np.arange(-1,5,0.001)

# 신호 x1과 x2 정의
x1 = ((t / 2) * (t / 2) * ((0 <= t) & (t < 2))) + (3.0 * ((2 <= t) & (t < 4)))
x2 = t * ((0 <= t) & (t < 2)) + -1 * ((2 <= t) & (t < 4))

fig,axes=plt.subplots(2,figsize=(6,6.5))
axes[0].plot(t,x1)
axes[1].plot(t,x2)
```

```python
# 문제: P1.22
# 학번: 201735030
# 이름: 이호민

t = np.arange(-1,3,0.005) # 시간 변수 정의
x = lambda t:(1 * ((0 <= t) & (t < 1))) + (0.5 * ((1 <= t) & (t < 2)))

# 신호 변형
x1 = lambda t:x(t+1) # 시간 이동
x2 = lambda t:x1(-t) # 반전
x3 = lambda t:x2(1/2 * t) # 확장

fig, axes = plt.subplots(2,3,figsize=(8,8))
fig.delaxes(axes[0,0])
fig.delaxes(axes[0,2])
axes[0,1].plot(t, x(t))
axes[0,1].set_title('$x(t)$')
axes[1,0].plot(t, x1(t))
axes[1,0].set_title('$x_1(t) = x(t+1)$')
axes[1,1].plot(t, x2(t))
axes[1,1].set_title('$x_2(t) = x_1(-t)$')
axes[1,2].plot(t, x3(t))
axes[1,2].set_title('$x_3(t) = x_2(1/2 * t)$')
```

```python
# 문제: P1.23
# 학번: 201735030
# 이름: 이호민

# 사각 함수 추가 메서드
def rect(t, t0 = 0, Tw = 1.0):
  return ((t - t0) / Tw)

x = lambda t: (1 * (abs(t) < 1 / 2)) + (0 * (abs(t) > 1/2))
x1 = lambda t: x(rect(t, 1))
x2 = lambda t: x1(rect(t, 0, 2))

t = np.arange(-2,5,0.001) # 시간 변수 정의
fig, axes = plt.subplots(2,2,figsize=(8,8))
fig.delaxes(axes[0,1])

axes[0,0].plot(t, x(t))
axes[0,0].set(title="$rect(t)$", xlim=[-1, 3])
axes[1,0].plot(t, x1(t))
axes[1,0].set(title="$rect(t - 1)$", xlim=[-1, 3])
axes[1,1].plot(t, x2(t))
axes[1,1].set(title="$rect((t - 1) / 2)$", xlim=[0, 4])
```

```python
# 문제: P1.24
# 학번: 201735030
# 이름: 이호민

t = np.arange(-0.1,0.5,0.001) # 시간 변수 정의
x = 2 * np.cos(np.pi * ((20 * t) + (1 / 3)))
fig, ax = plt.subplots()

ax.plot(t, x)
ax.set(title="$x(t) = A cos(\omega t + \phi)$", xlim=[-0.1, 0.5], xlabel="t(sec)")
```

```python
# 문제: P1.25
# 학번: 201735030
# 이름: 이호민

x = lambda t:np.cos(2 * np.pi * t) + np.sin((np.pi * t) - (np.pi / 3))
x1 = lambda t:(x(t) + x(-t)) / 2
x2 = lambda t:(x(t) + x(t)) / 2

t = np.arange(-2,2,0.001) # 시간 변수 정의
fig, axes = plt.subplots(2,2,figsize=(8,8))
fig.delaxes(axes[0,1])

axes[0,0].plot(t, x(t))
axes[0,0].set(title="$x(t)$", ylim=[-2,2])
axes[1,0].plot(t, x1(t))
axes[1,0].set(title="$x_e(t)$", xlabel="t(sec)", ylim=[-2,2])
axes[1,1].plot(t, x2(t))
axes[1,1].set(title="$x_o(t)$", xlabel="t(sec)", ylim=[-2,2])
```

```python
# 문제: P1.26A
# 학번: 201735030
# 이름: 이호민

import sympy as sp

x1, t1 = sp.symbols('x1 t1')
x2, t2 = sp.symbols('x2 t2')
x1 = 10 * cos(2 * np.pi * 10 * t1 + np.radians(45))
x2 = 5 * cos(2 * np.pi * 5 * t2 + np.radians(30))
print("결과: ", sp.integrate(x1**2,(t1,0,1)) + sp.integrate(x2**2,(t2,0,1)))
```
