```python
import numpy as np
import matplotlib.pyplot as plt

# 문제: P1.21
# 학번: 201735030
# 이름: 이호민

t = np.arange(-1,5,0.001)

# 신호 x1과 x2 정의
x1 = ((t / 2) * (t / 2) * ((0 <= t) & (t < 2))) + (3.0 * ((2 <= t) & (t < 4)))
x2 = t * ((0 <= t) & (t < 2)) + -1 * ((2 <= t) & (t < 4))

fig,axes=plt.subplots(2,figsize=(6,6.5))
axes[0].plot(t,x1)
axes[1].plot(t,x2)
```

```python
# 문제: P1.22
# 학번: 201735030
# 이름: 이호민

t = np.arange(-1,3,0.005) # 시간 변수 정의
x = lambda t:(1 * ((0 <= t) & (t < 1))) + (0.5 * ((1 <= t) & (t < 2)))

# 신호 변형
x1 = lambda t:x(t+1) # 시간 이동
x2 = lambda t:x1(-t) # 반전
x3 = lambda t:x2(1/2 * t) # 확장

fig, axes = plt.subplots(2,3,figsize=(8,8))
fig.delaxes(axes[0,0])
fig.delaxes(axes[0,2])
axes[0,1].plot(t, x(t))
axes[0,1].set_title('$x(t)$')
axes[1,0].plot(t, x1(t))
axes[1,0].set_title('$x_1(t) = x(t+1)$')
axes[1,1].plot(t, x2(t))
axes[1,1].set_title('$x_2(t) = x_1(-t)$')
axes[1,2].plot(t, x3(t))
axes[1,2].set_title('$x_3(t) = x_2(1/2 * t)$')
```
